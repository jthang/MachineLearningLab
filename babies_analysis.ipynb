```python
df['totalwgt_kg'] = df.totalwgt_lb / 2.2
totalwgt = df['totalwgt_kg'].value_counts().sort_index()
totalwgt.plot()

plt.xlim((0.0, 6))
plt.ylim((0.0, 350))
plt.xlabel('Weight of Baby (Kg)')
plt.ylabel('No. of Babies')
plt.title('Weight of Babies')
plt.legend(loc='best')
plt.tight_layout()

# titanic.groupby('pclass').survived.sum().plot(kind='bar')
```

```python
df['agepreg'].value_counts().sort_index().plot()

plt.xlim((0.0, 50))
plt.ylim((0.0, 100))
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age of Mummies')
plt.legend(loc='best')
plt.tight_layout()
```

```python
outcome = df.outcome.value_counts().sort_index()
outcome = outcome.rename(index = {
                            1 :'Live Birth',
                            2 :'Induced Abortion',
                            3 :'Still Birth',
                            4 :'Miscarriage',
                            5 :'Ectopic Pregnancy',
                            6 :'Current Pregnancy',
})

outcome.plot(kind='bar')

```
