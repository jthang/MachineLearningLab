#Movie Genres Exploratory Data Analysis

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all the ticks first
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #use this to re-activate the axis
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
#converting text files to dataframe
names = ['imdbID', 'title', 'year', 'score', 'votes', 'runtime', 'genres']
data = pd.read_csv('data/imdb_top_10000.txt', delimiter='\t', names=names).dropna()
data.head()
```

```python
#cleaning up the runtime to make it numeric
#run this experiment before using it as list of comprehension
dirty = '142 mins.'
number, text = dirty.split(' ')
clean = int(number)
print number
```

```python
clean_runtime = [float(r.split(' ')[0]) for r in data.runtime]
data['runtime'] = clean_runtime
data.head()
```

```python
data.genres.value_counts(sort=True)
```

```python
data.dtypes
```

```python
#split up the genre
genres = set() #create an empty set
for m in data.genres:
    genres.update(g for g in m.split('|'))
genres = sorted(genres)

#make a new column for each genre
for genre in genres:
    data[genre] = [genre in movie.split('|') for movie in data.genres]

data.head()
```

```python
#Run experiments to see until which character to remove year
data.title[0][:-7]
```

```python
#Removing year from the title
data['title'] = [t[0:-7] for t in data.title] #get all the characters except last 7
data.head()
```

```python
decade = (data.year // 10) * 10
tyd = data[['title', 'year']]
tyd['decade'] = decade
tyd.head()
```

```python
decade_mean = data.groupby(decade).score.mean()
decade_mean.name = 'Decade Mean'
print decade_mean
```

```python
plt.plot(decade_mean.index, decade_mean.values, 'o-',
         color='r', lw=3, label='Decade Average')
plt.scatter(data.year, data.score, alpha=.04, lw=0, color='k')
```

```python
grouped_scores = data.groupby(decade).score
mean = grouped_scores.mean()
std = grouped_scores.std()

plt.fill_between(decade_mean.index, (decade_mean + std).values,
                 (decade_mean - std).values, color='r', alpha=.2)
plt.plot(decade_mean.index, decade_mean.values, 'o-',
         color='r', lw=3, label='Decade Average')
plt.scatter(data.year, data.score, alpha=.04, lw=0, color='k')
```

```python
year_mean = data.groupby('year').score.mean()
std = year_mean.std()

plt.scatter(data.year, data.score, alpha=.04, lw=0, color='k')
plt.plot(year_mean.index, year_mean.values, '-', color='r', lw=1, label='Yearly Average')
plt.fill_between(year_mean.index, 
                (year_mean + std).values,
                (year_mean - std).values, 
                color='r', alpha=.2)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=6, figsize=(12, 8), 
                         tight_layout=True)

bins = np.arange(1950, 2013, 3)
for ax, genre in zip(axes.ravel(), genres):
    ax.hist(data[data[genre] == 1].year.values, 
           bins=bins, histtype='stepfilled', 
           normed=True, alpha=.3, ec='none')
    ax.hist(data.year, bins=bins, 
            histtype='stepfilled', ec='None', 
            normed=True, zorder=0, color='#cccccc')
    
    ax.annotate(genre, xy=(1955, 3e-2), fontsize=14)
    ax.xaxis.set_ticks(np.arange(1950, 2013, 30))
    ax.set_yticks([])
    remove_border(ax, left=False)
    ax.set_xlabel('Year')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=6, figsize=(12, 8), 
                         tight_layout=True)

bins = np.arange(30, 240, 10)
for ax, genre in zip(axes.ravel(), genres):
    ax.hist(data[data[genre] == 1].runtime.values, 
           bins=bins, histtype='stepfilled', 
           normed=True, alpha=.3, ec='none')
    ax.hist(data.runtime, bins=bins, 
            histtype='stepfilled', ec='None', 
            normed=True, zorder=0, color='#cccccc')
    
#     ax.set_xticks([])
    ax.set_yticks([])
    remove_border(ax, left=False)
    ax.set_xlabel('Runtime [min]')
    ax.annotate(genre, xy=(200, .02), fontsize=12)
```
