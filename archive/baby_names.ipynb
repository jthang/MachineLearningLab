```python
%matplotlib inline
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.mpl_style', 'default')
plt.rcParams['figure.figsize'] = (15, 5)

pd.set_option('html', False)
pd.set_option('max_columns', 30)
pd.set_option('max_rows', 20)
```

```python
#Separated by commas, so can just put into a dataframe
names1880 = pd.read_csv('data/names/yob1880.txt', names=['name', 'sex', 'births'])
names1880               
```

```python
#Get the sum of the births
names1880.groupby('sex').births.sum()
```

```python
#Get all the files into 1 dataframe
years = range(1880, 2011)
pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = 'data/names/yob%d.txt' % year
    frame = pd.read_csv(path, names = columns)
    
    frame['year'] = year #new column 'year' to indicate the year
    pieces.append(frame)

names = pd.concat(pieces, ignore_index=True)
names[:10]
```

```python
#Create a pivot table
total_births = names.pivot_table('births', index = 'year',
                                 cols = 'sex', aggfunc = sum)
total_births[:10]
```

```python
total_births.plot()
```

```python
#create a function for the prop, takes in argument group
def add_prop(group):
    births = group.births.astype(float) #make all the births a floating number
    
    group['prop'] = births / births.sum() #new column 'prop' which is probability
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)
names[:10]
```

```python
#check the prop columns add up to 1
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
#function to get top1000 names for each sex/year
def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000
```

```python
#splitting top1000 names to boys and girls
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
#pivot table no. of births by year and name
total_births = top1000.pivot_table('births', index = 'year', cols = 'name', aggfunc=sum)
total_births
```

```python
#create a subset to find out different names
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
#create subplots
subset.plot(subplots=True, figsize=(12,10), grid=False,
            title='No.of births per year')
```

```python
#Measure the diversity over the years
table = top1000.pivot_table('prop', index='year', cols='sex', aggfunc=sum)
table[:10]
```

```python
#shows there's increasing diversity (drop in prop)
table.plot()
```

```python
#find number of popular names in top 50%
#consider for boy names from 2010 first
df = boys[boys.year == 2010] #select all the index which column is 2010
df[:10]
```

```python
#use cumulative sum to find how long it takes for prop to reach 0.5
#sort index by prop column (descending to start with most popular), then cumsum
prop_cumsum = df.sort_index(by='prop', ascending=False).prop.cumsum()
prop_cumsum[:10]
```

```python
#how many it takes to get 50% in 2010?
prop_cumsum.values.searchsorted(0.5)
```

```python
#how many it takes to get 50% in 1900?
df1900 = boys[boys.year == 1900]
prop_cumsum = df1900.sort_index(by='prop', ascending=False).prop.cumsum()
prop_cumsum.values.searchsorted(0.5)
```

```python
#get for both boys and girls in different years
#define function to get the index to reach 50$
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q)

#apply it to the top1000 group, group by year, and sex
diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity[:10]
```

```python
#unstack to get F and M to be columns
diversity.unstack()
```

```python
#no. of popular names in top 50%
diversity.plot()
```

```python
#find the last letter of all the boys names
#aggregate all the births by year, sex, and final letter
#get last letter from name column
get_last_letter = lambda x: x[-1] #create a lambda function to extract last letter
last_letters = names.name.map(get_last_letter) #apply lambda function to 'name' column in 'names' dataframe
last_letters.name = 'last_letter' #last_letters index name is last_letter
last_letters[:10]
```

```python
#create a pivot table based on births data
table = names.pivot_table('births', index=last_letters,
                          cols=['sex', 'year'], aggfunc=sum)
table[:10]
```

```python
#create a subtable of 3 representative years
#at the level of "year", select only 3 columns
subtable = table.reindex(columns = [1910, 1960, 2010], level='year')
subtable.head()
```

```python
#sum of all the births in the subtable
subtable.sum()
```

```python
#get proportion of last_letter to sum of births
#divide entire table by the total births
letter_prop = subtable / subtable.sum().astype(float)
letter_prop.head()
```

```python
#compare the last letter proportions over the years
letter_prop['M'].plot(kind='bar', rot=0)
```

```python
letter_prop['F'].plot(kind='bar', rot=0)
```

```python
#get proportion of letter_prop to the whole table
letter_prop = table / table.sum().astype(float)
dny_ts = letter_prop.ix[['d', 'n', 'y'], 'M'].T #extract the index dny and transpose as columns
dny_ts.head()
```

```python
dny_ts.plot()
```

```python
#boy names become girl names (vice versa)
all_names = top1000.name.unique() #get only the unique names
mask = np.array(['lesl' in x.lower() for x in all_names]) #select all the names which begins with lesl
lesley_like = all_names[mask]
lesley_like
```

```python
#filter out the names from top1000 and sum them
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.head()
```

```python
#group all the names together and sum them
filtered.groupby('name').births.sum()
```

```python
#pivot table to sort by M and F
table = filtered.pivot_table('births', index='year', cols='sex',
                             aggfunc=sum)
table.head()
```

```python
#whole table divide by the table sum (proportion)
table_prop = table.div(table.sum(axis=1), axis=0)
table_prop
```

```python
table_prop.plot(style={'M':'k-', 'F':'k--'})
```
