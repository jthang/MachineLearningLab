###<span style="background-color:#66FF99)">Generate Data</span>

```python
%pylab inline
import numpy as np
# import matplotlib as plt

def f(x):
    return x * np.sin(x) + np.sin(2 * x)

def gen_data(n_samples=200):
    np.random.seed(15)
    X = np.random.uniform(0, 10, size=n_samples)[:, np.newaxis]
    y = f(X.ravel()) + np.random.normal(scale=2, size=n_samples)
    train_mask = np.random.randint(0, 2, size=n_samples).astype(np.bool)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = gen_data(100)

x_plot = np.linspace(0, 10, 500)

FIGSIZE = (11, 7)

def plot_data(alpha=0.4, s=20):
    fig = plt.figure(figsize=FIGSIZE)
    gt = plt.plot(x_plot, f(x_plot), alpha=alpha, label='ground truth')

    # plot training and testing data
    plt.scatter(X_train, y_train, s=s, alpha=alpha)
    plt.scatter(X_test, y_test, s=s, alpha=alpha, color='red')
    plt.xlim((0, 10))
    plt.ylabel('y')
    plt.xlabel('x')
    
plot_data()
```

###<span style="background-color:#66FF99)">Fit Decision Regression Tree</span>

```python
#Fit regression tree
from sklearn.tree import DecisionTreeRegressor
plot_data()
est = DecisionTreeRegressor(max_depth=1).fit(X_train, y_train)
plt.plot(x_plot, est.predict(x_plot[:, np.newaxis]),
         label='RT max_depth=1', color='g', alpha=0.9, linewidth=2)

est = DecisionTreeRegressor(max_depth=5).fit(X_train, y_train)
plt.plot(x_plot, est.predict(x_plot[:, np.newaxis]),
         label='RT max_depth=5', color='purple', alpha=0.9, linewidth=1)

plt.legend(loc='upper left')
```

###<span style="background-color:#66FF99)">Add Gradient Boosting</span>

```python
#Add gradient boosting to regression trees
from itertools import islice
plot_data()

est = GradientBoostingRegressor(n_estimators=1000, max_depth=1, learning_rate=1.0)
est.fit(X_train, y_train)

ax = plt.gca()
first = True

for pred in islice(est.staged_predict(x_plot[:, np.newaxis]), 0, est.n_estimators, 10):
    plt.plot(x_plot, pred, color='r', alpha=0.2)

pred = est.predict(x_plot[:, np.newaxis])
plt.plot(x_plot, pred, color='g')
```

###<span style="background-color:#66FF99)">Deviance Plot</span>
* To determine if model is overfitting
* Plots the training/test error as function of number of trees (i.e complexity)
* Training error stored in es.train_score_
* Test error computed with est.staged_predict

```python
def deviance_plot(est, X_test, y_test, ax=None, label='', train_color='#2c7bb6', 
                  test_color='#d7191c', alpha=1.0, ylim=(0, 10)):
    n_estimators = len(est.estimators_)
    test_dev = np.empty(n_estimators)
    
    for i, pred in enumerate(est.staged_predict(X_test)):
        test_dev[i] = est.loss_(y_test, pred)
    
    if ax is None:
        fig = plt.figure(figsize=FIGSIZE)
        ax = plt.gca()
        
    ax.plot(np.arange(n_estimators) + 1, test_dev, color=test_color, label='Test %s' % label, 
            linewidth=2, alpha=alpha)
    ax.plot(np.arange(n_estimators) + 1, est.train_score_, color=train_color, 
            label='Train %s' % label, linewidth=2, alpha=alpha)
    ax.set_ylabel('Error')
    ax.set_xlabel('n_estimators')
    ax.set_ylim(ylim)
    return test_dev, ax
        
test_dev, ax = deviance_plot(est, X_test, y_test)
ax.legend(loc='upper right')
        
```

###<span style="background-color:#66FF99)">Regularization</span>
* Large gap between train and test error shows overfitting
* Need to control overfitting using Regularization
* 3 methods:
    1. Tree structure
    * Shrinkage
    * Stochastic Gradient Boosting (SGB)

###<span style="background-color:#66FF99)">1. Tree Structure</span>
* **max_depth** of trees: control degree of features interactions (variance++)
* **min_samples_leaf**: indicate no. of samples per leaf (bias++)

```python
def fmt_params(params):
    return ", ".join("{0}={1}".format(key, val) for key, val in params.iteritems())

fig = plt.figure(figsize=FIGSIZE)
ax = plt.gca()
for params, (test_color, train_color) in [({}, ('#d7191c', '#2c7bb6')),
                                          ({'min_samples_leaf': 3}, #IMPT!
                                           ('#fdae61', '#abd9e9'))]:
    est = GradientBoostingRegressor(n_estimators=1000, max_depth=1, 
                                    learning_rate=1.0)
    est.set_params(**params)
    est.fit(X_train, y_train)
    test_dev, ax = deviance_plot(est, X_test, y_test, ax=ax, label=fmt_params(params),
                                 train_color=train_color, test_color=test_color)

plt.legend(loc='upper right')
```

###<span style="background-color:#66FF99)">2. Shrinkage</span>
* **learning_rate**: slow learning by shrinking predictions of each tree by some small scalar
* low learning requires high n_estimators

```python
fig = plt.figure(figsize=FIGSIZE)
ax = plt.gca()
for params, (test_color, train_color) in [({}, ('#d7191c', '#2c7bb6')),
                                          ({'learning_rate': 0.1}, #IMPT!
                                           ('#fdae61', '#abd9e9'))]:
    est = GradientBoostingRegressor(n_estimators=1000, max_depth=1, 
                                    learning_rate=1.0)
    est.set_params(**params)
    est.fit(X_train, y_train)
    test_dev, ax = deviance_plot(est, X_test, y_test, ax=ax, label=fmt_params(params),
                                 train_color=train_color, test_color=test_color)

plt.legend(loc='upper right')
    
```

###<span style="background-color:#66FF99)">3. Stochastic Gradient Boosting</span>
* Subsampling the training set before growing each tree (**subsample**)
* Subsampling the features before finding the best split node (**max_features**)
* Using use max_features more often than subsamples when large no. of features!

```python
fig = plt.figure(figsize=FIGSIZE)
ax = plt.gca()
for params, (test_color, train_color) in [({}, ('#d7191c', '#2c7bb6')),
                                          ({'learning_rate': 0.1,
                                            'subsample': 0.5}, #IMPT!
                                           ('#fdae61', '#abd9e9'))]:
    est = GradientBoostingRegressor(n_estimators=1000, max_depth=1, 
                                    learning_rate=1.0)
    est.set_params(**params)
    est.fit(X_train, y_train)
    test_dev, ax = deviance_plot(est, X_test, y_test, ax=ax, label=fmt_params(params),
                                 train_color=train_color, test_color=test_color)

plt.legend(loc='upper right')
```

###<span style="background-color:#66FF99)">Hyperparameter Tuning</span>
**USE THE FOLLOWING RECIPE TO TUNE THE HYPERPARAMETERS**
1. Pick ``n_estimators`` as large as possible computationally (e.g 3000)
* Use **grid search** to tune ``max_depth``, ``learning_rate``, ``min_sample_leaf``, and ``max_features``
* Increase ``n_estimators`` even more and tune ``learning_rate`` again, holding other parameters fixed

* All parameters interact with each other, so be careful when tuning the parameters

```python
from sklearn.grid_search import GridSearchCV

param_grid = {'learning_rate': [0.1, 0.01, 0.001],
              'max_depth': [4, 6],
              'min_samples_leaf': [3, 5]  ## depends on the nr of training examples
              # 'max_features': [1.0, 0.3, 0.1] ## not possible in our example (only 1 fx)
              }
est = GradientBoostingRegressor(n_estimators=3000)
gs_cv = GridSearchCV(est, param_grid, scoring='mean_squared_error', n_jobs=4).fit(X_train, y_train)

print ('Best Hyperparameters: %r' % gs_cv.best_params_)
```

```python
#refit the model using the best parameters
est.set_params(**gs_cv.best_params_)
est.fit(X_train, y_train)

#plot the approximation
plot_data()
plt.plot(x_plot, est.predict(x_plot[:, np.newaxis]), color='r', linewidth=2)
```
