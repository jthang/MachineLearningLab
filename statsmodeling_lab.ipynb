#Method of Moments

```python
df_mean = df.mean()
df_mean
```

```python
df_var = df.var()
df_var
```

```python
alpha_mom = df_mean ** 2 / df_var
beta_mom = df_var / df_mean
estimates = pd.DataFrame({'alpha_mom': alpha_mom, 'beta_mom': beta_mom})
estimates
```

```python
from scipy.stats import gamma

xvals = np.linspace(0, 10)
yvals = gamma.pdf(xvals, alpha_mom[0], beta_mom[0])
df.Jan.hist(normed=True, bins=20)
plt.plot(xvals, yvals)
```

```python
axs = df.hist(normed=True, figsize=(12,8), sharex=True, sharey=True,
              bins=15, grid=False)

for ax in axs.ravel():
    m = ax.get_title()
    x = np.linspace(*ax.get_xlim())
    ax.plot(x, gamma.pdf(x, alpha_mom[m], beta_mom[m]))
    
    label = 'alpha = {0:.2f}\nbeta = {1:.2f}'.format(alpha_mom[m], beta_mom[m])
    ax.annotate(label, xy=(10, 0.2))
plt.tight_layout()
```
