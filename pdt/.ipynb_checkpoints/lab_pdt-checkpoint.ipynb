### Data Cleaning

```python
# Price

df['special_price'] = df['special_price'] / 10000
df['original_price'] = df['original_price'] / 10000
df['disc_price'] = df['special_price']
df.loc[df['disc_price'].isnull(), 'disc_price'] = df['original_price']

df[['original_price', 'special_price', 'disc_price']].head()
```

### Feature Engineering

```python
data1 = "disc_price" #continuous
data2 = "stock" #continuous

sns.jointplot(data1, data2, df, kind="reg");
```

```python
data = df.net_sale_count #continuous

plt.hist(data1, bins=25, color="#6495ED", histtype="stepfilled", normed=False, alpha=.5)
plt.xlabel("net_sale_count")
plt.ylabel("Frequency")
sns.despine();
```

```python
# Discount Price

df['disc_pct'] = (1 - (df['disc_price']/df['original_price'])) * 100

# Calculate number of days products have launched

from datetime import datetime
now = datetime.now()

df['activated_at'] = pd.to_datetime(df['activated_at'])
df['days_launch'] = (now - df.activated_at).astype('timedelta64[D]')

# Comparing sales and impression relationships

df['sale_imp_30'] = (df.net_sale_count_last_30days - df.net_sale_count_last_7days) / (df.impressions_count_last_30days - df.impressions_count_last_7days) * 100
df['sale_imp_total'] = (df.net_sale_count - df.net_sale_count_last_7days) / (df.impressions_count - df.impressions_count_last_7days) * 100

new_features = ['disc_pct', 'days_launch', 'sale_imp_30', 'sale_imp_total']
df[new_features].describe()
```

```python
# Downsize to few features

features = ['colors', 'gender', 'season_group', 'brand', 'original_price',
            'disc_price', 'disc_pct', 'days_launch', 'sale_imp_30', 'sale_imp_total']
target = ['net_sale_count_last_7days']

df_clean = df[features]
print df_clean.shape
df_clean.head()
```

```python
data = df_clean.disc_price #continuous

plt.hist(data, bins=250, color="#6495ED", histtype="stepfilled", alpha=.5)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.xlim(0, 100)
sns.despine()
plt.figure();

data1 = df_clean.days_launch #Continuous

plt.hist(data1, bins=20, color="#6495ED", histtype="stepfilled", normed=False, alpha=.5)
plt.xlabel("Days Launched")
plt.ylabel("Frequency")
sns.despine();
```
