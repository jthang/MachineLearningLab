```python
#specific list of columns to extract
info_keys = ['description', 'group', 'id', 'manufacture']
info = DataFrame(db, columns=info_keys)
info[:5]
```

```python
group_count.plot(kind='barh')
```

```python
#Put all into 1 dataframe, including ID of food
nutrients = []
for rec in db:
    fnuts = DataFrame(rec['nutrients'])
    fnuts['id'] = rec['id']
    nutrients.append(fnuts)
nutrients = pd.concat(nutrients, ignore_index=True)
```

```python
#see rows with columns 'id' = 1008
nutrients.loc[nutrients['id'] == 1008] 
```

```python
nutrients
```

```python
#check how many duplicates
nutrients.duplicated().sum()
```

```python
#drop all duplicates
nutrients = nutrients.drop_duplicates()
nutrients_count = pd.value_counts(nutrients.group)
nutrients_count[:10]
```

```python
#change the name of the columns for Info
col_mapping = {'description':'food',
               'group':'fgroup'}
info = info.rename(columns=col_mapping, copy=False)
info
```

```python
#change the name of the columns for Nutrient
col_mapping = {'description':'nutrient',
               'group':'nutgroup'}
nutrients = nutrients.rename(columns=col_mapping, copy=False)
nutrients
```

```python
#merge Info and Nutrients
ndata = pd.merge(nutrients, info, on='id', how='outer')
ndata
```

```python
#group by nutrient, fgroup, and value
result = ndata.groupby(['nutrient','fgroup'])['value'].quantile(0.5)
result
```

```python
#Sort the results in Zinc to see which food contains the most Zinc
result['Zinc, Zn'].order()
```

```python
#plot the bar chart for zinc
result['Zinc, Zn'].order().plot(kind='barh')
```

```python
#find which food is most dense in each nutrient
by_nutrient = ndata.groupby(['nutgroup','nutrient'])
get_maximum = lambda x: x.xs(x.value.idxmax())
get_minimum = lambda x: x.xs(x.value.idxmin())

max_foods = by_nutrient.apply(get_maximum)[['value','food']]
max_foods.food = max_foods.food.str[:50]
```
