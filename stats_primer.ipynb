```python
% pylab inline
import scipy as sp
import pandas
```

###Plotting Waves

```python
t = arange(0,10,0.1)

plot(t, sin(t))
plot(t, cos(t))
```

###Generating Data

```python
#Using standard y = mx + c equation
x = arange(-100,100)
par = {'m':0.5, 'c':100, 'noise':4}
y = par['m']*x + par['c'] + par['noise']*sp.randn(len(x))

xHigh = x[where(t>10)]
yHigh = y[where(t>10)]

plot(xHigh, yHigh)
```

###Getting the best-fit line

```python
#Generating the matrix
xMatrix = vstack((xHigh, ones(len(xHigh)))).T
#Use linear algebra least squares to find the slope and intercept
slope, intercept = linalg.lstsq(xMatrix, yHigh)[0]

plot(xHigh, yHigh)
plot(xHigh, slope*xHigh + intercept, 'g')

print "Slope: %.5f, and intercept: %.5f" % (slope, intercept)
```
