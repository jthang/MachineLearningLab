#List of Countries Population

```python
url = 'http://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population'
r = requests.get(url)
print r
soup = BeautifulSoup(r.content) #use content instead of text
tables = soup.find("table", {"class":['sortable', 'wikitable']})
```

```python
population_tables = []
for table in soup.find_all('table', {'class': ['sortable', 'wikitable']}):
    if 'sortable' in table.attrs.get('class') and 'wikitable' in table.attrs.get('class'):
        if table.find('th').text == 'Country or territory':
            population_tables.append(table)
population_tables
```

```python
column_names = []
col_indexes = []
table_head = table.find_all('th')
print table_head

for idx, col_name in enumerate(table_head):
    if col_name.text != '%':
        col_indexes.append(idx)
        column_names.append(col_name.text)
# print column_names
# print col_indexes

rows_data = []
table_row = table.find_all('tr')
for idx, row_td in enumerate(table_row):
        row = []
        for col_idx, col_val in enumerate(row_td.find_all('td')):
            if col_idx in col_indexes:
                if col_idx == 0:
                    row.append(col_val.text[1:])
                else:
                    row.append(int(col_val.text.replace(',', '')))
        if row:
            rows_data.append(row)
print rows_data
```

```python
population_df = []
for table in population_tables:
    table_df = pd.DataFrame(rows_data, columns=column_names)
    table_df = table_df.set_index(['Country or territory'])
    table_df = table_df.applymap(np.int32)
print table_df.shape
table_df.head()
```

#Singapore Economic History

```python
def extract(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
#     table = soup.find('span', {"id": "Economic_history"}).findNext('table', {"class": "wikitable"})
    table = soup.find('table', {"class": "wikitable"})
    
    economy = list()
    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if len(cells) == 5:
            year = cells[0].string
            gdp = cells[1].string
            usd = cells[2].string
            nominal = cells[3].string
            ppp = cells[4].string
            economy.append({'year':year, 
                         'gdp':gdp,
                         'usd':usd,
                         'nominal':nominal})
        else: pass
    return pd.DataFrame(economy)
        
df = extract("http://en.wikipedia.org/wiki/Economy_of_Singapore")
indexed_df = df.set_index(['year'])
indexed_df
```

#International Trade Agreements

```python
def trade(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
#     table = soup.find('span', {"id": "Economic_history"}).findNext('table', {"class": "wikitable"})
    table = soup.find('table', {"class": "wikitable sortable"})
    
    int_trade = list()
    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if len(cells) == 7:
            economy = cells[0].string
            agreement = cells[1].string
            concluded = cells[3].string
            signed = cells[4].string
            effective = cells[5].string
            int_trade.append({'economy':economy, 
                              'agreement':agreement,
                              'concluded':concluded,
                              'effective':effective})
        else: pass
    return pd.DataFrame(int_trade)
        
df = trade("http://en.wikipedia.org/wiki/Economy_of_Singapore")
indexed_df = df.set_index(['economy'])
indexed_df
```

#American Presidents

```python
SITE = "https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States"

def get_date(cells):
    date = cells.find('span', {'class': 'date'})
    if date is not None: date = str(date.string)
    else: pass
    if date: date = datetime.strptime(date.strip(), "%B %d, %Y")
    return date

def pres(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    table = soup.find('table', {"class": "wikitable"})
    
    pres_list = list()
    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if len(cells) > 4:
            pres= cells[1].findAll('a')[0].string
            took_office = get_date(cells[2])
            left_office = get_date(cells[3])
            party= cells[4].findAll('a')[0].string
            pres_list.append({'pres':pres, 
                              'took_office':took_office,
                              'left_office':left_office,
                              'party':party})
        else: pass
    return pd.DataFrame(pres_list)
        
df = pres(SITE)
indexed_df = df.set_index(['pres'])
indexed_df.sort_index()
```
